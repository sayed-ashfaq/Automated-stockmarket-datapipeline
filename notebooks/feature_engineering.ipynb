```python
## Objective: Get the raw data from Yfinance and store it in csv file in data folder
import yfinance as yf
import os
```

```python
US_tickers=  ['TSLA', 'META', 'NVDA', 'PLTR', 'INTC']
indian_tickers = ['TATAMOTORS.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'ICICIBANK.NS']
```

```python
# fetch just one ticker to test
data = yf.download('AAPL', period= '3y', interval= '1d')
data
```

```python
# get indian stock data
ind_data = yf.download('TATAMotors.NS', period= '3y', interval= '1d')
ind_data
```

```python
# get all the data of just 5 stocks each from US and India and store it in data folder
data_folder = '../data/raw/us_stocks'
os.makedirs(data_folder, exist_ok=True)
for ticker in US_tickers:
    data = yf.download(ticker, period= '3y', interval= '1d')
    data.to_csv(os.path.join(data_folder, f'{ticker}.csv'))
```

```python
data_folder = '../data/raw/indian_stocks'
os.makedirs(data_folder, exist_ok=True)
for ticker in indian_tickers:
    data = yf.download(ticker, period= '3y', interval= '1d')
    data.to_csv(os.path.join(data_folder, f'{ticker}.csv'))
```
